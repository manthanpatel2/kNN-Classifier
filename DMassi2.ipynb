```python
import csv
import random
from collections import Counter
import matplotlib.pyplot as plt
```

```python
#READ THE DATA FROM FILE
def read_iris_dataset(filename):
    f = open(filename, "r")
    iris_dataset = list(csv.reader(f))
    iris_dataset.pop(-1)
    f.close()
    
    return iris_dataset

#MODIFY THE DATA    
def modify_iris_data(data):
    for row in data:
        for i in range(4):
            row[i] = float(row[i].strip())  

    return assign_iris_class(data)

def assign_iris_class(data):
    iris_classes = [row[4] for row in data]
    classes = dict()
    for i, value in enumerate(set(iris_classes)):
        classes[value] = i
    for row in data:
        row[4] = classes[row[4]]

    return data,classes
```

```python
raw_data = read_iris_dataset("iris.data") 
X,y = modify_iris_data(raw_data)
```

```python
#Euclidean distance calculation
def Euclidean(row1,row2): 
    Edistanceance = 0.0
    for i in range(len(row1)-1):
        Edistanceance += (row1[i]-row2[i])**2
    return (Edistanceance)**(1/2)

print( 'Euclidean distanceance：',Euclidean((5.0,3.4,1.5,0.2),(5.5,3.5,1.3,0.2)))

#Normalidisted_Euclidean calculation
def Normalidisted_Euclidean(row1,row2):   
    Ndistanceance = 0
    for i in range(len(row1)-1):
        mean = (row1[i]-row2[i])/2
        d = ( (row1[i] - mean) ** 2 + (row2[i] - mean) ** 2 )**(1/2)
        Ndistanceance += ((row1[i]-row2[i])/d ) ** 2
    return (Ndistanceance)**(1/2)

print( 'Normalidisted Euclidean distanceance：',Normalidisted_Euclidean((1,1),(7,2)))

#Cosine_Similarity calculation
def Cosine_Similarity(row1,row2):   
    distance = 0.0
    x,y = 0,0
    for i in range(len(row1)-1):
        distance += row1[i]*row2[i]
        x += row1[i]**2 
        y += row2[i]**2 
        similarity = distance/( (x*y)**(1/2)) 
        
    return (1-similarity)

print( 'Cosine Similarity：',Cosine_Similarity((8,5,1,1),(1,2,5,9)))


```

```python
#B-Implement kNN


def kNN_e_distance(data, num_neighbors):
    correct_result = 0
    for point in data:
        tempdata = data.copy()
        tempdata.remove(point)
        
        dist = []
        for d_row in tempdata:
            distance = Euclidean(point, d_row)
            dist.append((d_row, distance))
        dist.sort(key=lambda tup: tup[1])
        nebrs = []
        for i in range(num_neighbors):
            nebrs.append(dist[i][0])
        
        nebrs_array = []
        for nebr in nebrs:
            nebrs_array.append(nebr[-1])
        counter = Counter(nebrs_array)
        pred_array = counter.most_common(1)[0][0]
        if pred_array == point[-1]:
            correct_result += 1

    return correct_result

def kNN_ne_distance(data, num_neighbors):
    correct_result = 0
    for point in data:
        tempdata = data.copy()
        tempdata.remove(point)
        
        dist = []
        for d_row in tempdata:
            distance = Normalidisted_Euclidean(point, d_row)
            dist.append((d_row, distance))
        dist.sort(key=lambda tup: tup[1])
        nebrs = []
        for i in range(num_neighbors):
            nebrs.append(dist[i][0])
        
        nebrs_array = []
        for nebr in nebrs:
            nebrs_array.append(nebr[-1])
        counter = Counter(nebrs_array)
        pred_array = counter.most_common(1)[0][0]
        if pred_array == point[-1]:
            correct_result += 1

    return correct_result

def kNN_cos_distance(data, num_neighbors):
    correct_result = 0
    for point in data:
        tempdata = data.copy()
        tempdata.remove(point)
        
        dist = []
        for d_row in tempdata:
            distance = Cosine_Similarity(point, d_row)
            dist.append((d_row, distance))
        dist.sort(key=lambda tup: tup[1])
        nebrs = []
        for i in range(num_neighbors):
            nebrs.append(dist[i][0])
        
        nebrs_array = []
        for nebr in nebrs:
            nebrs_array.append(nebr[-1])
        counter = Counter(nebrs_array)
        pred_array = counter.most_common(1)[0][0]
        if pred_array == point[-1]:
            correct_result += 1

    return correct_result
    
```

```python
#C-Calculating accuracy
num_of_neighbor = [1,3,5,7]
Euclidean_array = []
for k in num_of_neighbor:  
    accuracy_e = kNN_e_distance(development, k)/len(development)
    Euclidean_array.append(accuracy_e*100)
    #print('Accuracy for Number of neighbor', k,'is',accuracy_e )
    #print(accuracy_e)

Cosine_array = []
for k in num_of_neighbor:
    accuracy_c = kNN_cos_distance(development, k)/len(development)
    Cosine_array.append(accuracy_c*100)
    #print('Accuracy for Number of neighbor', k,'is',accuracy_c )
    #print(accuracy_c)

Normalized_array = []
for k in num_of_neighbor:
    accuracy_n = kNN_ne_distance(development, k)/len(development)
    Normalized_array.append(accuracy_n*100)
    #print('Accuracy for Number of neighbor', k,'is',accuracy_n )
    #print(accuracy_n)

print('Euclidean distance:',Euclidean_array)
print('\nCosine similarity:',Cosine_array)
print('\nNormalized euclidean distance:',Normalized_array)
```
